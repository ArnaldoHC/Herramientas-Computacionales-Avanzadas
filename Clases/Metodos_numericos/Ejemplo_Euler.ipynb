Aproximar la soluci贸n del problema de valor inicial  $\frac{dy}{dt}=2y-1$ con condici贸n inicial $y(0)=1$ con $\Delta t=0.1$ sobre el intevalo $0\leq t \leq 1$.
Identificamos que la ecuaci贸n $f(t,y)=f(y)=2y-1$ y discretizamos la ecuaci贸n diferencial de la siguiente forma.
$$y(t_{n+1})=y(t_n)+\left[2\,y(t_n)-1\right]\Delta t$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Soluci贸n num茅rica de la ecuaci贸n diferencial dy/dt = 2y - 1
# Intervalo de tiempo
dt = 0.1  # par谩metro que refina el modelo cuando dt->0
t = np.arange(0, 1 + dt, dt)  # valores del tiempo espaciados con el intervalo dt

# Condici贸n inicial
y0 = 1  # Puedes cambiar esta condici贸n inicial seg煤n sea necesario
y = np.zeros_like(t)  # inicializaci贸n del vector soluci贸n en ceros
y[0] = y0  # condici贸n inicial para el tiempo t=0

# M茅todo de Euler para resolver la ecuaci贸n diferencial
for i in range(1, len(t)):
    y[i] = y[i-1] + dt * (2 * y[i-1] - 1)

# Graficar la soluci贸n num茅rica
plt.figure()
plt.plot(t, y, 'b-o', label='Soluci贸n num茅rica')
plt.title('Soluci贸n num茅rica de dy/dt = 2y - 1')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)

# Soluci贸n anal铆tica de la ecuaci贸n diferencial
# y(t) = C*exp(2t) + 0.5
# donde C es una constante determinada por la condici贸n inicial
C = (y0 - 0.5)
y_analitica = C * np.exp(2 * t) + 0.5

# Graficar la soluci贸n anal铆tica
plt.plot(t, y_analitica, 'r--', label='Soluci贸n anal铆tica')
plt.legend()
plt.show()

# Calcular el error entre la soluci贸n num茅rica y la soluci贸n anal铆tica
error = np.abs(y - y_analitica)

plt.figure()
plt.plot(t, error, 'k-o')
plt.title('Error entre la soluci贸n num茅rica y anal铆tica')
plt.xlabel('t')
plt.ylabel('Error')
plt.grid(True)
plt.show()
```

#### Ahora calcuemos el intervalo $\Delta t$ tal que el $error<0.1$
