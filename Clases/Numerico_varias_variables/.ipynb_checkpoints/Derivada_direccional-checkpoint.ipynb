# Derivada direccional
La derivada direccional de una función $f(x,y,z)$ en la dirección de un vector unitario $u = (u_1, u_2, u_3)$ se define como:

$$D_uf = \nabla f \cdot u = \frac{\partial f}{\partial x}u_1 + \frac{\partial f}{\partial y}u_2 + \frac{\partial f}{\partial z}u_3$$

Donde $\nabla f$ es el gradiente de $f$ y $u=(u_1,u_2,u_3)$ un vector unitario que determina la dirección del cambio sobre la función 

```python
import numpy as np
from numpy import sqrt, arctan2

def derivada_direccional(f, punto, direccion, h=1e-5):
    """
    Calcula la derivada direccional usando diferencias finitas centradas
    
    Parámetros:
        f: función escalar f(x,y,z)
        punto: (x0,y0,z0) donde evaluar
        direccion: vector dirección [u1,u2,u3] (no necesita ser unitario)
        h: tamaño del paso
        
    Retorna:
        Valor de la derivada direccional
    """
    # Normalización del vector dirección
    u = np.array(direccion, dtype=float)
    u = u/np.linalg.norm(u)
    
    # Cálculo del gradiente por diferencias finitas centradas
    x0, y0, z0 = punto
    grad = np.zeros(3)
    
    # Derivada parcial en x
    grad[0] = (f(x0+h, y0, z0) - f(x0-h, y0, z0))/(2*h)
    # Derivada parcial en y
    grad[1] = (f(x0, y0+h, z0) - f(x0, y0-h, z0))/(2*h)
    # Derivada parcial en z
    grad[2] = (f(x0, y0, z0+h) - f(x0, y0, z0-h))/(2*h)

    # Producto punto entre el gradiente y el vector unitario
    return np.dot(grad, u) 
```

## Ejemplo.
Calcular la tasa de cambio de la temperatura en un punto del manto terrestre en dirección a un punto de posible pluma mantélica. Usa un model modelo térmico simplificado para la temperatura en °C:

$$T(x,y,z)=20−0.5z+0.01z^2+0.1xy$$

```python
# Función de temperatura
def temperatura_manto(x, y, z):
    return 20 - 0.5*z + 0.01*z**2 + 0.1*x*y

# Punto y dirección
punto = (100, 100, -500)
direccion = (30, 40, -10)

# Cálculo
dd = derivada_direccional(temperatura_manto, punto, direccion)
print(f"Derivada direccional de la temperatura: {dd:.2f} °C/km")

# Visualización
x_vals = np.linspace(0, 200, 50)
y_vals = np.linspace(0, 200, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = -500 * np.ones_like(X)  # Plano a z=-500

T = temperatura_manto(X, Y, Z)

plt.figure(figsize=(12, 6))

# Gráfica 2D
plt.subplot(1, 2, 1)
contour = plt.contourf(X, Y, T, levels=20, cmap='magma')
plt.colorbar(contour, label='Temperatura (°C)')
plt.quiver(punto[0], punto[1], direccion[0], direccion[1], 
           color='white', scale=10, width=0.005)
plt.title('Mapa de temperatura a 500km profundidad')
plt.xlabel('Coordenada Este (km)')
plt.ylabel('Coordenada Norte (km)')

# Gráfica 3D
ax = plt.subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(X, Y, T, cmap='magma', alpha=0.8)
ax.scatter(punto[0], punto[1], temperatura_manto(*punto), 
           color='red', s=100, label='Punto de estudio')
ax.set_title('Superficie de temperatura 3D')
ax.set_xlabel('Este (km)')
ax.set_ylabel('Norte (km)')
ax.set_zlabel('Temperatura (°C)')
plt.colorbar(surf, ax=ax, label='Temperatura (°C)')
plt.tight_layout()
plt.show()
```

## Ejemplo
Calcular la tasa de cambio de la densidad de plasma en la magnetosfera terrestre en dirección al viento solar. Utilizando el modelo de densidad de plasma:

$$n(x,y,z)=10^6\cdot e^{-r/7}\cdot (2+\text{sen}\theta)$$

donde $n$ es el número de partículas por unidad de volumen en $cm^3$ y la distancia $r=\sqrt{x^2+y^2+z^2}$ y $\theta=\left(\frac{\sqrt{y^2+z^2}}{x}\right)$

```python
# Función de densidad de plasma
def densidad_plasma(x, y, z):
    r = sqrt(x**2 + y**2 + z**2)
    theta = arctan2(sqrt(y**2 + z**2), x)
    return 1e6 * np.exp(-r/7) * (2 + np.sin(theta))

# Punto y dirección
punto = (10, 0, 0)
direccion = (0.9, 0.1, -0.05)

# Cálculo
dd_plasma = derivada_direccional(densidad_plasma, punto, direccion)
print(f"Derivada direccional: {dd_plasma:.2e} partículas/cm³ por RT")

# Visualización
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 50)
THETA, PHI = np.meshgrid(theta, phi)
R = 10  # Radio de visualización

# Convertir a cartesianas
X = R * np.sin(PHI) * np.cos(THETA)
Y = R * np.sin(PHI) * np.sin(THETA)
Z = R * np.cos(PHI)

# Calcular densidad en la superficie
D = densidad_plasma(X, Y, Z)

fig = plt.figure(figsize=(14, 6))

# Gráfica 3D
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, facecolors=plt.cm.plasma(D/np.max(D)),
                       rstride=1, cstride=1, alpha=0.8)
ax1.scatter(punto[0], punto[1], punto[2], color='red', s=100)
ax1.quiver(punto[0], punto[1], punto[2], 
           direccion[0], direccion[1], direccion[2],
           length=3, color='white', arrow_length_ratio=0.1)
ax1.set_title('Densidad de plasma en la magnetosfera')
ax1.set_xlabel('X (RT)')
ax1.set_ylabel('Y (RT)')
ax1.set_zlabel('Z (RT)')

# Corte 2D
ax2 = fig.add_subplot(122)
y_vals = np.linspace(-15, 15, 100)
z_vals = np.linspace(-15, 15, 100)
Y, Z = np.meshgrid(y_vals, z_vals)
X = 10 * np.ones_like(Y)
D_corte = densidad_plasma(X, Y, Z)

contour = ax2.contourf(Y, Z, D_corte, levels=20, cmap='plasma')
plt.colorbar(contour, label='Densidad (partículas/cm³)')
ax2.set_title('Corte en X=10 RT')
ax2.set_xlabel('Y (RT)')
ax2.set_ylabel('Z (RT)')

plt.tight_layout()
plt.show()
```
